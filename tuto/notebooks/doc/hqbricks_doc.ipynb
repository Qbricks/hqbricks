# Hqbricks documentation

Partial documentation of Hqbricks for the tutorial.

## Prog

Note that the prog type is often referred as circuit in the exercises as they are a subsed of what we call programs.

### pr_int

Int type of prog.

#### `~$int_val`

##### Description
Unary operator to construct a constant pr_int from an int value.

##### Parameters
- `int_val` (`int`): The int value of the constant pr_int.

##### Returns
`pr_int`: Constant pr_int of value int_val.

#### `ivar var_name`

##### Description
Construct a variable pr_int with the given name.

##### Parameters
- `var_name` (`string`): Name of the variable pr_int.

##### Returns
`pr_int`: Variable pr_int of name var_name.

### pr_bool

Bool type of prog.

#### `qbit_val qreg i`

##### Description
Get the $i_{th}$ element of qreg as a pr_bool.

##### Parameters
- `qreg` (`qreg`): The qreg to index.
- `i` (`pr_int`): The index.

##### Returns
`pr_bool`: Bool value of the $i_{th}$ element of qreg.

#### `cbit_val creg i`

##### Description
Get the $i_{th}$ element of creg as a pr_bool.

##### Parameters
- `creg` (`creg`): The creg to index.
- `i` (`pr_int`): The index.

##### Returns
`pr_bool`: Bool value of the $i_{th}$ element of creg.

#### `!bool`

##### Description
Not operator.

##### Parameters
- `bool` (`pr_bool`): The bool.

##### Returns
`pr_bool`: Not bool value.

#### `bool1 & bool2`

##### Description
And operator.

##### Parameters
- `bool1` (`pr_bool`): The first bool.
- `bool2` (`pr_bool`): The second bool.

##### Returns
`pr_bool`: bool1 and bool2 value.

### Gate

#### `h`

##### Description
The Hadamard gate.

#### `x`

##### Description
The X (not) gate.

#### `rz a`

##### Description
The Z-Rotation gate of angle a.

##### Parameters
`pr_int`: The angle.

#### `z`

##### Description
Shortcut for Rz 1.

### Register

#### `qreg qreg_name len`

##### Description
Construct a qreg with the given name and length.

##### Parameters
- `qreg_name` (`string`): Name of the qreg.
- `len` (`pr_int`): Length of the qreg.

##### Returns
`qreg`: Qreg with the given name and length.

#### `creg creg_name len`

##### Description
Construct a creg with the given name and length.

##### Parameters
- `creg_name` (`string`): Name of the creg.
- `len` (`pr_int`): Length of the creg.

##### Returns
`creg`: Creg with the given name and length.

#### `q_idx qreg i`

##### Description
Get the $i_{th}$ element of qreg.

##### Parameters
- `qreg` (`qreg`): The qreg to index.
- `i` (`pr_int`): The index.

##### Returns
`qreg`: Subset of the qreg containing only the $i_{th}$ element.

### Program instructions

#### `init_qreg qreg`

##### Description
Init the qreg with 0 everywhere.

##### Parameters
- `qreg` (`qreg`): The qreg to init.

##### Returns
`prog`: The initialization of qreg.

#### `prog1 -- prog2`

##### Description
Sequence operator.

##### Parameters
- `prog1` (`prog`): The first program of the sequence.
- `prog2` (`prog`): The second program of the sequence.

##### Returns
`prog`: The sequence of prog1 and prog2

#### `qreg |> gate`

##### Description
Gate application operator, apply gate on qreg.

##### Parameters
- `qreg` (`qreg`): The qreg, if it contains multiple qubits the gate is applied parallely on all the qregs.
- `gate` (`gate`): The gate to apply.

##### Returns
`prog`: The application of gate on qreg.

#### `bool => prog`

##### Description
Control operator.

##### Parameters
- `bool` (`pr_bool`): The control condition.
- `prog` (`prog`): The controlled prog.

##### Returns
`prog`: If bool do prog.

#### `qreg -@ creg`

##### Description
Measure operator.

##### Parameters
- `qreg` (`qreg`): The qreg to measure.
- `creg` (`creg`): The target creg.

##### Returns
`prog`: Measure of qreg in creg.

## HPS

### Evaluation

#### `evaluate_prog ?(rewrite_settings = no_rewrite) ?(print = true) prog hps`

##### Description
Evaluate a prog into an HPS.

##### Parameters
- `?rewrite_settings`: Rewrite settings, `all_auto` for all automatic rewrites (Change-Var and HH), `interactive_rewrite` for interactive rewrite (doesn't work inside Jupyter).
- `?print` (`bool`): Whether to print information about the evaluation and rewriting or not.
- `prog` (`prog`): The prog to evaluate.
- `hps` (`Hps.t`): The input HPS in which we evaluate prog.

##### Returns
`Hps.t`: The evaluated HPS.

### Interactive rewrite

Here are some available commands during interactive rewrite.

#### `hh [-find|-all] y0 y1`

##### Description
HH rewrite rule, with no options it tries to apply HH with y0 and y1 and outputs an explaining error if it cannot.  
If an option is given, parameters are not needed.

##### Options
- `-find`: Find all y pairs where HH can be applied.
- `-all`: Find and apply all possible HH.

##### Parameters
- `y0`: First y.
- `y1`: Second y.

#### `change_var [-all] y0 "y0 + x0"`

##### Description
Change-Var rewrite rule, for now it only works to add a constant or an input variable to a path variable.  
If an option is given, parameters are not needed.

##### Options
- `-all`: Find and apply all Change-Var.

##### Parameters
- `y0`: Path variable to substitute.
- `"y0 + x0"`: Boolean expression to replace y0.

#### `discard q alice ...`

##### Description
Try to discard the given registers (at least 1 is needed).

##### Parameters
- `q alice ...`: The registers to discard

#### `exit`

##### Description
Exit the interactive rewrite.

## Assertion

#### `hps_eq hps1 hps2`

##### Description
Check if hps1 and hps2 are equal (without transformations).

##### Parameters
- `hps1` (`Hps.t`): The first HPS.
- `hps2` (`Hps.t`): The second HPS.

##### Returns
`bool`: Whether or not hps1 and hps2 are equal.

#### `hps_satisfies spec_hps hps`

##### Description
Check if hps satisfies spec_hps (with potential automatic discard).

##### Parameters
- `spec_hps` (`Hps.t`): The specification HPS.
- `hps` (`Hps.t`): The HPS.

##### Returns
`bool`: Whether or not hps satisfies spec_hps.

#### `hps_satisfies_proba_qmem p spec_qmem hps`

##### Description
Check if the probability that hps satisfies spec_qmem is p.

##### Parameters
- `p` (`Hps.Scalar.t`): The expected probability.
- `spec_qmem` (`Hps.Mem.t`): The specification qmem.
- `hps` (`Hps.t`): The HPS.

##### Returns
`bool`: Whether or not the probability that hps satisfies spec_qmem is p.

#### `unitary_eq prog1 prog2 input_hps`

##### Description
Check if unitary circuits prog1 and prog2 are equal with the same input_hps.

##### Parameters
- `prog1` (`prog`): The first prog.
- `prog2` (`prog`): The second prog.
- `input_hps` (`Hps.t`): The input HPS.

##### Returns
`bool`: Whether or not prog1 and prog2 are equal with the same input_hps.

## Concretization

#### `Vector_map.of_hps hps`

##### Description
Compute the Vector Map of hps.

##### Parameters
- `hps` (`Hps.t`): The HPS.

##### Returns
`Vector_map.t`: The resulting Vector Map.

#### `hps_proba_qmem spec_qmem hps`

##### Description
Compute the probability that hps satisfies spec_qmem.

##### Parameters
- `spec_qmem` (`Hps.Mem.t`): The specification qmem.
- `hps` (`Hps.t`): The HPS.

##### Returns
`Hps.Scalar.t`: The probability that hps satisfies spec_qmem.
