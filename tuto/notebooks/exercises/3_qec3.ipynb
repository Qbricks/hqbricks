# 3-qubit error correction

## Introduction

The QEC3 algorithm is an error correction algorithm that encodes 1 logical qubit into 3 physical qubits to protect against single bit-flip errors.

In this exercise, we will use the QEC3 algorithm to:
- Implement an hybrid circuit with a different gate set
- Evaluate the circuit into an HPS and rewrite using the HH rule
- Concretize the HPS into a Vector Map and compute some probabilities
- Verify the probability of valid output
- Generalize the example by using a symbolic probability of error $p$

- The double controls are controls of the form if $c_0$ **and** $c_1$ then ...

Since there are 3 $y$ path variables, there are 8 possible paths (1 for each possible combination of values of the $y$), and we also have 8 worlds (8 distinct possible classical memory), so in this case we have 1 world per path, meaning that there was no interference.  
Here, we can compute the probability of having no error $\ket{x_0}_\psi$ $\ket{x_0x_0}_q$ with the **hps_proba_qmem** function with a qmem spec $\ket{x_0}_\psi$ $\ket{x_0x_0}_q$ as parameter:

```ocaml
let x0 = Hps.Hket.of_var @@ X 0
let qmem_spec = Hps.Mem.(Stdlib.(empty |> add ("psi", 0) x0 |> add ("q", 0) x0 |> add ("q", 1) x0))
let success_proba = hps_proba_qmem qmem_spec hps

let () = print_endline "\nQmem Spec:"
let _ = Jupyter_notebook.display "text/latex" ("\\[" ^ Hps.Mem.qmem_to_latex (qmem_spec) ^ "\\]")

let () = print_endline "\nSuccess Proba:"
let _ = Jupyter_notebook.display "text/latex" ("\\[" ^ Hps.Scalar.to_latex (success_proba) ^ "\\]")
```

## Verification

Knowing the success probability in advance, we can verify the success probability of the HPS with:

```ocaml
let exp_proba = Hps.Scalar.(SFrac (Z.of_int 7, Z.of_int 27)) in
assert (Assertion.hps_satisfies_proba_qmem exp_proba qmem_spec hps);
print_endline "Specification verified"
```

You may have noticed that for $p = \frac{2}{3}$ the probability of success of QEC3 is lesser than the initial probability of success $\frac{1}{3}$, but the QEC3 becomes better and better as $p$ decreases. ## Symbolic Probability

We now generalize the the previous example by replacing the numeric probability of error with a symbolic value $p$.

```ocaml
let input_hps = Hps.add_qmem_vec_x ("psi", 0) 1 0 Hps.one

let hps =
  let p = Hps.Scalar.(SVar "p") in
  let qec3_circ = generate_qec3 p in
  Evaluator.(evaluate_prog qec3_circ input_hps ~rewrite_settings:all_auto ~print:(Pr_none))

let x0 = Hps.Hket.of_var @@ X 0
let qmem_spec = Hps.Mem.(Stdlib.(empty |> add ("psi", 0) x0 |> add ("q", 0) x0 |> add ("q", 1) x0))
let success_proba = hps_proba_qmem qmem_spec hps

let _ = Jupyter_notebook.display "text/latex" ("Input HPS:" ^ Hps.to_latex_block input_hps)
let _ = Jupyter_notebook.display "text/latex" ("HPS:" ^ Hps.to_latex_block hps)

let () = print_endline "\nQmem Spec:\n:"
let _ = Jupyter_notebook.display "text/latex" ("\\[" ^ Hps.Mem.qmem_to_latex (qmem_spec) ^ "\\]")
  
let () = print_endline "\nSuccess Proba:"
let _ = Jupyter_notebook.display "text/latex" ("\\[" ^ Hps.Scalar.to_latex (success_proba) ^ "\\]")
```
