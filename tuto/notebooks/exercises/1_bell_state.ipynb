# Bell state

## Introduction

In this exercise, we will recreate a classic example of quantum entanglement - the **Bell state** - using our tool, **Hqbricks**.

This will give you a first hands-on look at how to:
- Build a **quantum circuit**
- Evaluate it into a **Hybrid Path-Sum (HPS)**, the mathematical representation we use to analyze and verify hybrid circuits
- Concretize the HPS into a **Vector Map** to look at the entangled state

We will do this twice, first with an **input initialized** to $\ket{00}_q$ , and then with a **symbolic input** $\ket{x_0x_1}_q$ .

## Simulation

In this part we will implement and analyze the Bell state circuit with an input initialized to $\ket{00}_q$.

### Circuit implementation

To load the **Hqbricks** library, run the following code:

```ocaml
#require "hqbricks"
open Hqbricks
```

Now, we will define the Bell state circuit using several operators:
- `--` for sequencing operations
- `|>` for gate applications
- **=>** for quantum controll

Additionaly, notice that the integers in circuit description must be written using the unary operator `~$`, for example `~$2`.

```ocaml
open Gate_set_impl.Clifford_k (* Import the gate set containing h, x, z, rz, swap *)

let bell_circ =
  let open Prog in (* Import the functions and operators needed to construct circuits *)
  let q = qreg "q" ~$2 in (* Declare a quantum register with 2 qubits named "q" *)
  init_qreg q (* Initialize q to |00> *)
  -- (q_idx q ~$0 |> h) (* Apply the Hadamard gate to q at index 0 *)
  -- (qbit_val q ~$0 => (q_idx q ~$1 |> x)) (* Apply the X (NOT) gate to q at index 1 controlled by the qubit value of q at index 0 *)

let () =
  print_endline "\nCirc:";
  Prog.print bell_circ
```

### HPS evaluation

In this step, we will evaluate the circuit into a HPS using the **Evaluator.evaluate_prog** function, the input HPS is one as we don't need to set any input value since they are all initialized by the circuit.

```ocaml
let input_hps = Hps.one
let () =
  print_endline "\nInput HPS:";
  print_endline (Hps.to_string input_hps);
  print_endline ""

let hps = Evaluator.evaluate_prog bell_circ input_hps
let () =
  print_endline "\nHPS:";
  print_endline (Hps.to_string hps)
```

### Vector Map concretization

After evaluating the circuit into a HPS, we can concretize the HPS into a Vector Map using the **Vector_map.of_hps** function and visualize the output quantum state.

```ocaml
open Concretization

let vm = Vector_map.of_hps hps
let () =
  print_endline "\nVector Map:";
  print_endline (Vector_map.to_string vm)
```

As expected, we obtain the entangled quantum state: $\frac{\ket{00} + \ket{11}}{\sqrt{2}}$.
